```python
# 画像ダウンロード
from urllib.request import urlretrieve

urlretrieve('http://uta.pw/shodou/img/28/214.png', '214.png')
```

```python
# OpenCVで画像読み込み
import cv2

img = cv2.imread('214.png')
print(img)
```

```python
# 画像を画面に表示
import matplotlib.pyplot as plt
%matplotlib inline

#plt.axis('off') # 座標軸非表示
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 画像のリサイズ
img2 = cv2.resize(img, (600, 300))
# 画像の一部切り取り [y, x]
img3 = img[150:450, 100:300]
# リサイズした画像を保存
cv2.imwrite('214-resize.png', img2)
cv2.imwrite('214-cut.png', img3)
# 画像を表示
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
plt.show()
```
